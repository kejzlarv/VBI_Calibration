# Modules

```python
# Required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec

import pickle as pkl
```

# Simulation - Scalability

```python
#### This reproduces the scalability plots from pre-computed results
fig, axes = plt.subplots(nrows=3, ncols=1, figsize = (7.5,7))
plt.subplots_adjust(hspace = 0.001)
plt.rcParams.update({'font.size': 16})
lw = 2

markevery = 1000
############## 2.5K
n_obs = 2500
n_steps = 200000
S = 50
l = 5
vine_type = "D"
learning_rate = "AdaGrad"
eta = 0.07
decay = 1.0

time_steps_MH = np.load("Simulation_results/Calibration/time_nobs_2500_2500_nsample_2000_sampler_MH.npy")
time_MH = (time_steps_MH - time_steps_MH[0]) / 3600
with open('Simulation_results/Calibration/2500_MH' + '.pickle', 'rb') as handle:
    pred_results_MH = pkl.load(handle)
time_steps_NUTS = np.load("Simulation_results/Calibration/time_nobs_2500_2500_nsample_1000_sampler_NUTS.npy")
time_NUTS = (time_steps_NUTS - time_steps_NUTS[0]) / 3600
with open('Simulation_results/Calibration/2500_NUTS' + '.pickle', 'rb') as handle:
    pred_results_NUTS = pkl.load(handle)

folder = str(n_obs) + "_S_" + str(S) + "_l_" + str(l) + "_vine_" + vine_type + "_learning_" + learning_rate + \
                                    "_eta_" + str(eta) + "_decay_" + str(decay)

time_steps = np.load("Simulation_results/Calibration/OBB_" + folder + "/" + "time_nobs_" + str(n_obs) + "_S_" + str(S) + "_l_" + str(l) +  "_vine_" + vine_type + "_step_" + str(n_steps) +".npy")
time_VI = (time_steps - time_steps[0]) / 3600
with open("Simulation_results/Calibration/OBB_" + folder + '.pickle', 'rb') as handle:
    pred_results = pkl.load(handle)

axes[0].plot(time_VI,pred_results['mse'], label = r"VC", color = "red", lw = lw, ls = ':',markevery=markevery)
axes[0].plot(time_MH,pred_results_MH['mse'], label = "MH", color = "darkorange", lw = lw, ls = '-.')
axes[0].plot(time_NUTS,pred_results_NUTS['mse'], label = "NUTS", color = "darkgreen", lw = lw, ls = '-')
axes[2].set_xlim([0,6])
axes[2].set_ylim([0.002,0.2])
#axes[2].set_xticklabels(())
axes[2].set_yscale('log')


############## 5K
n_obs = 5000
n_steps = 200000
S = 50
l = 5
vine_type = "D"
learning_rate = "AdaGrad"
eta = 0.06
decay = 1.0

time_steps_MH = np.load("Simulation_results/Calibration/time_nobs_5000_5000_nsample_500_sampler_MH.npy")
time_MH = (time_steps_MH - time_steps_MH[0]) / 3600
with open('Simulation_results/Calibration/5000_MH' + '.pickle', 'rb') as handle:
    pred_results_MH = pkl.load(handle)
time_steps_NUTS = np.load("Simulation_results/Calibration/time_nobs_5000_5000_nsample_150_sampler_NUTS.npy")
time_NUTS = (time_steps_NUTS - time_steps_NUTS[0]) / 3600
with open('Simulation_results/Calibration/5000_NUTS' + '.pickle', 'rb') as handle:
    pred_results_NUTS = pkl.load(handle)

folder = str(n_obs) + "_S_" + str(S) + "_l_" + str(l) + "_vine_" + vine_type + "_learning_" + learning_rate + \
                                    "_eta_" + str(eta) + "_decay_" + str(decay)

time_steps = np.load("Simulation_results/Calibration/OBB_" + folder + "/" + "time_nobs_" + str(n_obs) + "_S_" + str(S) + "_l_" + str(l) +  "_vine_" + vine_type + "_step_" + str(n_steps) +".npy")
time_VI = (time_steps - time_steps[0]) / 3600
with open("Simulation_results/Calibration/OBB_" + folder + '.pickle', 'rb') as handle:
    pred_results = pkl.load(handle)

axes[1].plot(time_VI,pred_results['mse'], label = r"VC", color = "red", lw = lw, ls = ':',markevery=markevery)
axes[1].plot(time_MH,pred_results_MH['mse'], label = "MH", color = "darkorange", lw = lw, ls = '-.')
axes[1].plot(time_NUTS,pred_results_NUTS['mse'], label = "NUTS", color = "darkgreen", lw = lw, ls = '-')
axes[0].set_xlim([0,6])
axes[0].set_ylim([0.002,0.2])
axes[0].set_xticklabels(())
axes[0].set_yscale('log')


############## 10K
n_obs = 10000
n_steps = 150000
S = 50
l = 5
vine_type = "D"
learning_rate = "AdaGrad"
eta = 0.09
decay = 1.0

time_steps_MH = np.load("Simulation_results/Calibration/time_nobs_10000_10000_nsample_80_sampler_MH.npy")
time_MH = (time_steps_MH - time_steps_MH[0]) / 3600
with open('Simulation_results/Calibration/10000_MH' + '.pickle', 'rb') as handle:
    pred_results_MH = pkl.load(handle)
time_steps_NUTS = np.load("Simulation_results/Calibration/time_nobs_10000_10000_nsample_80_sampler_NUTS.npy")
time_NUTS = (time_steps_NUTS - time_steps_NUTS[0]) / 3600
with open('Simulation_results/Calibration/10000_NUTS' + '.pickle', 'rb') as handle:
    pred_results_NUTS = pkl.load(handle)

folder = str(n_obs) + "_S_" + str(S) + "_l_" + str(l) + "_vine_" + vine_type + "_learning_" + learning_rate + \
                                    "_eta_" + str(eta) + "_decay_" + str(decay)

time_steps = np.load("Simulation_results/Calibration/OBB_" + folder + "/" + "time_nobs_" + str(n_obs) + "_S_" + str(S) + "_l_" + str(l) +  "_vine_" + vine_type + "_step_" + str(n_steps) +".npy")
time_VI = (time_steps - time_steps[0]) / 3600
with open("Simulation_results/Calibration/OBB_" + folder + '.pickle', 'rb') as handle:
    pred_results = pkl.load(handle)

axes[2].plot(time_VI,pred_results['mse'], label = r"VC", color = "red", ls = ':', lw = lw,markevery=markevery)
axes[2].plot(time_MH,pred_results_MH['mse'], label = "MH", color = "darkorange", ls = '-.', lw = lw)
axes[2].plot(time_NUTS,pred_results_NUTS['mse'], label = "NUTS", color = "darkgreen", ls = '-', lw = lw)
axes[1].set_xlim([0,6])
axes[1].set_ylim([0.002,0.2])
axes[1].set_yscale('log')
axes[2].set_xlabel('Time [h]')
axes[1].set_xticklabels(())
# Additional setting
fig.text(0.01, 0.5, 'MSE', va='center', rotation='vertical')
axes[1].annotate(r"$n = 1 \times 10^4$", xy=(.07, .75), xycoords=axes[1].transAxes, color = "black")
axes[0].annotate(r"$n =0.5 \times 10^4$", xy=(.01, .75), xycoords=axes[0].transAxes, color = "black")
axes[2].annotate(r"$n = 2 \times 10^4$", xy=(.07, .75), xycoords=axes[2].transAxes, color = "black")
axes[1].legend(frameon=False, bbox_to_anchor=(0.5,0.55), ncol = 2)
plt.savefig("Scalability_3.pdf", dpi=300,bbox_inches='tight')
plt.show()
```

# Memory Profiles

```python
### Loading memory profiles data - MCMC
# 5K
df_25MH = pd.read_csv("Simulation_results/Memory_profiles/mprofile_2500_MH.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_25MH_ind = df_25MH.iloc[:,0].values != '__main__.VC_calibration_hyper_RB'
df_25MH_mem = df_25MH.iloc[df_25MH_ind,0].values
df_25MH_mem = df_25MH_mem[1:].astype(float)

df_25MH_time = df_25MH.iloc[df_25MH_ind,1].values
df_25MH_time = df_25MH_time[1:].astype(float)
df_25MH_time = (df_25MH_time - df_25MH_time[0])

# 5K
df_25N = pd.read_csv("Simulation_results/Memory_profiles/mprofile_2500_NUTS.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_25N_ind = df_25N.iloc[:,0].values != '__main__.VC_calibration_hyper_RB'
df_25N_mem = df_25N.iloc[df_25N_ind,0].values
df_25N_mem = df_25N_mem[1:].astype(float)

df_25N_time = df_25N.iloc[df_25N_ind,1].values
df_25N_time = df_25N_time[1:].astype(float)
df_25N_time = (df_25N_time - df_25N_time[0])

# 10K
df_5MH = pd.read_csv("Simulation_results/Memory_profiles/mprofile_5000_MH.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_5MH_ind = df_5MH.iloc[:,0].values != '__main__.VC_calibration_hyper_RB'
df_5MH_mem = df_5MH.iloc[df_5MH_ind,0].values
df_5MH_mem = df_5MH_mem[1:].astype(float)

df_5MH_time = df_5MH.iloc[df_5MH_ind,1].values
df_5MH_time = df_5MH_time[1:].astype(float)
df_5MH_time = (df_5MH_time - df_5MH_time[0])

# 10K
df_5N = pd.read_csv("Simulation_results/Memory_profiles/mprofile_5000_NUTS.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_5N_ind = df_5N.iloc[:,0].values != '__main__.VC_calibration_hyper_RB'
df_5N_mem = df_5N.iloc[df_5N_ind,0].values
df_5N_mem = df_5N_mem[1:].astype(float)

df_5N_time = df_5N.iloc[df_5N_ind,1].values
df_5N_time = df_5N_time[1:].astype(float)
df_5N_time = (df_5N_time - df_5N_time[0])


# 20K
df_10MH = pd.read_csv("Simulation_results/Memory_profiles/mprofile_10000_MH.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_10MH_ind = df_10MH.iloc[:,0].values != '__main__.VC_calibration_hyper_RB'
df_10MH_mem = df_10MH.iloc[df_10MH_ind,0].values
df_10MH_mem = df_10MH_mem[1:].astype(float)

df_10MH_time = df_10MH.iloc[df_10MH_ind,1].values
df_10MH_time = df_10MH_time[1:].astype(float)
df_10MH_time = (df_10MH_time - df_10MH_time[0])

# 20K
df_10N = pd.read_csv("Simulation_results/Memory_profiles/mprofile_10000_NUTS.dat", 
                 index_col=0,
                 header=None) 

df_10N = df_10N.iloc[1:]
mem = []
tim = []
for i in range(len(df_10N)):
    mem = mem + [df_10N.index[i].split()[1]]
    tim = tim + [df_10N.index[i].split()[2]]
    
del(df_10N)
df_10N = pd.DataFrame(data = {"col1": mem, "col2":tim})  

df_10N_ind = df_10N.iloc[:,0].values != '__main__.VC_calibration_hyper_RB'
df_10N_mem = df_10N.iloc[df_10N_ind,0].values
df_10N_mem = df_10N_mem[1:].astype(float)

df_10N_time = df_10N.iloc[df_10N_ind,1].values
df_10N_time = df_10N_time[1:].astype(float)
df_10N_time = (df_10N_time - df_10N_time[0])
```

```python
### Loading memory profiles data - VC
# 5K
df_25VC = pd.read_csv("Simulation_results/Memory_profiles/mprofile_2500.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_25VC_ind = df_25VC.iloc[:,0].values != '__main__.VC_calibration_hyper_RB_CV_OD_simple'
df_25VC_mem = df_25VC.iloc[df_25VC_ind,0].values
df_25VC_mem = df_25VC_mem[1:].astype(float)

df_25VC_time = df_25VC.iloc[df_25VC_ind,1].values
df_25VC_time = df_25VC_time[1:].astype(float)
df_25VC_time = (df_25VC_time - df_25VC_time[0])

# 10K
df_5VC = pd.read_csv("Simulation_results/Memory_profiles/mprofile_5000.dat", 
                 sep="\s+", #separator whitespace
                 index_col=0,
                 header=None) 
df_5VC_ind = df_5VC.iloc[:,0].values != '__main__.VC_calibration_hyper_RB_CV_OD_simple'
df_5VC_mem = df_5VC.iloc[df_5VC_ind,0].values
df_5VC_mem = df_5VC_mem[1:].astype(float)

df_5VC_time = df_5VC.iloc[df_5VC_ind,1].values
df_5VC_time = df_5VC_time[1:].astype(float)
df_5VC_time = (df_5VC_time - df_5VC_time[0])

# 20K
df_10VC = pd.read_csv("Simulation_results/Memory_profiles/mprofile_10000.dat", 
                 index_col=0,
                 header=None) 

df_10VC = df_10VC.iloc[1:]
mem = []
tim = []
for i in range(len(df_10VC)):
    mem = mem + [df_10VC.index[i].split()[1]]
    tim = tim + [df_10VC.index[i].split()[2]]
    
del(df_10VC)
df_10VC = pd.DataFrame(data = {"col1": mem, "col2":tim})  

df_10VC_ind = df_10VC.iloc[:,0].values != '__main__.VC_calibration_hyper_RB_CV_OD_simple'
df_10VC_mem = df_10VC.iloc[df_10VC_ind,0].values
df_10VC_mem = df_10VC_mem[1:].astype(float)

df_10VC_time = df_10VC.iloc[df_10VC_ind,1].values
df_10VC_time = df_10VC_time[1:].astype(float)
df_10VC_time = (df_10VC_time - df_10VC_time[0])
```

```python
##### Ploting the loaded memory profiles
fig, axes = plt.subplots(nrows=3, ncols=1, figsize = (8,10))
plt.subplots_adjust(hspace = 0.001)
plt.rcParams.update({'font.size': 15})
lw = 2

markevery = 1000
divider_time = 60
divider_mem = 1024

axes[0].plot(df_25VC_time / divider_time, df_25VC_mem / divider_mem, label = r"$0.5 \times 10^4$", color = "red", lw = lw, ls = '-',markevery=markevery)
axes[0].plot(df_25MH_time/ divider_time, df_25MH_mem/ divider_mem, label = "MH", color = "darkorange", lw = lw, ls = '-.')
axes[0].plot(df_25N_time/ divider_time, df_25N_mem/ divider_mem, label = "NUTS", color = "darkgreen", lw = lw, ls = '--')
axes[0].set_xlim([0,60])
axes[0].set_ylim([0.002,18])
axes[0].set_xticklabels(())

axes[1].plot(df_5VC_time / divider_time, df_5VC_mem / divider_mem, label = r"VC", color = "red", lw = lw, ls = '-',markevery=markevery)
axes[1].plot(df_5MH_time / divider_time, df_5MH_mem / divider_mem, label = "MH", color = "darkorange", lw = lw, ls = '-.')
axes[1].plot(df_5N_time / divider_time, df_5N_mem / divider_mem, label = "NUTS", color = "darkgreen", lw = lw, ls = '--')
axes[1].set_xlim([0,60])
axes[1].set_ylim([0.002,70])
axes[1].set_xticklabels(())

axes[2].plot(df_10VC_time / divider_time, df_10VC_mem / divider_mem, label = r"VC", color = "red", ls = '-', lw = lw,markevery=markevery)
axes[2].plot(df_10MH_time / divider_time, df_10MH_mem / divider_mem, label = "MH", color = "darkorange", ls = '-.', lw = lw)
axes[2].plot(df_10N_time / divider_time, df_10N_mem / divider_mem, label = "NUTS", color = "darkgreen", ls = '--', lw = lw)
axes[2].set_xlim([0,60])
axes[2].set_ylim([0.002,300])
axes[2].set_xlabel('Time [min]')
# Additional setting
fig.text(0.01, 0.5, 'Memory [GB]', va='center', rotation='vertical')
axes[2].annotate(r'$n=2 \times 10^4$', xy=(.7, .65), xycoords=axes[2].transAxes, color = "black")
axes[0].annotate(r'$n=0.5 \times 10^4$', xy=(.7, .65), xycoords=axes[0].transAxes, color = "black")
axes[1].annotate(r'$n=1 \times 10^4$', xy=(.7, .65), xycoords=axes[1].transAxes, color = "black")
axes[1].legend(frameon=False, bbox_to_anchor=(0.1,0.1), ncol=3)
#plt.savefig("Scalability_memory_V2.pdf", dpi=300,bbox_inches='tight')
plt.show()
```
